```python
def create_features(matches: pd.DataFrame) -> pd.DataFrame:
    cols_to_use = []
    for col in matches.columns:
        if (col[0:1]=='h_'):
            col_name = col[2:] + '_diff'
            # Now need to go back a round and find the differences4
            matches[col_name] = matches[col] - matches['a_' + col[2:]]
        if (col[-6:] == '_h_ave'):
            col_name = col[:-6] + '_ave_diff'
            matches[col_name] = matches[col] - matches[col[:-6] + '_a_ave']
    return matches
```
