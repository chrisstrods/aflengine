```python
# Change date type so we can get everything in order
def change_date_type(matches: pd.DataFrame) -> pd.DataFrame:
    """
    Change date data to be type datetime
    """
    matches['date'] = pd.to_datetime(matches['date'],format="%d-%b-%Y")
    return matches
```

```python
merged = change_date_type(merged)
```

```python
df_sorted = merged.sort_values(by=['date'])
```

```python
df = df_sorted
```

```python
# df is now sorted. Add a column to store games played for each player.
df['player_gp'] = 0
```

```python
df.playerid
```

```python
def add_games_played(matches):
    games_played_dict = {}
    for idx,row in matches.iterrows():
        if row['playerid'] not in games_played_dict:
            games_played_dict[row['playerid']] = 1
            matches.at[idx,'player_gp'] = 1
        else:
            games_played_dict[row['playerid']] += 1
            matches.at[idx,'player_gp'] = games_played_dict[row['playerid']]
            
    return matches
```

```python
df = add_games_played(df)
```

```python
# Now we have the amount of games played at the player level. However, we want to see what the number of games
# played for the entire team is. We can use aggregate functions to achieve this.
```

```python
players['team']
```

```python
def games_played_per_team(matches,df):
    matches['h_games_played'] = 0
    matches['a_games_played'] = 0
    # Create the grouped dataframe with the aggregate of games played
    aggregation = df[['matchid','team','player_gp']].groupby(['matchid','team']).sum()
    aggregation.reset_index(inplace=True)
    # Find the relevant row to get the games played
    for idx,row in matches.iterrows():
        matches.at[idx,'h_games_played'] = aggregation[(aggregation['matchid'] == row['matchid']) &
                                                      (aggregation['team'] == row['hteam'])]['player_gp']
        matches.at[idx,'a_games_played'] = aggregation[(aggregation['matchid'] == row['matchid']) &
                                                      (aggregation['team'] == row['ateam'])]['player_gp']
    return matches
```

```python
matches = games_played_per_team(matches,df)
```
