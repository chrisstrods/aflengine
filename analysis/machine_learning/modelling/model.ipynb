```python
matches = pd.read_csv(matches_path)
```

```python
# Add a winners column to the matches file
def add_winner_col(matches):
    matches['winner'] = 0
    for idx,row in matches.iterrows():
        if row['hscore'] > row['ascore']:
            matches.at[idx,'winner'] = 1
        elif row['hscore'] == row['ascore']:
            matches.at[idx,'winner'] = .5
        else:
            matches.at[idx,'winner'] = 0
    return matches
```

```python
matches = add_winner_col(matches)
```

```python
# Now add that column to the features df
df['winner'] = matches['winner']
```
