For each of these players, we need to take into account their performance in their previous game, i.e. we need to add this to their statistics for the upcoming game.

```python
cols_to_update = [col for col in df.columns if (col[0:6] in ['career','season'])]
cols_to_update.remove('career_games_played')
cols_to_update.remove('season_games_played')
cols_to_update.remove('season')
```

```python
def update_last_game(df):
    for idx,row in df.iterrows():
        df.at[idx,'career_games_played'] += 1
        df.at[idx,'season_games_played'] += 1
        
        for col in cols_to_update:
            single_game_col = col[7:] # This is the non-aggregate column, e.g. 'Supercoach' instead of 'career_Supercoach'
            if col[0:7] == 'career_':
                df.at[idx,col] = (df.at[idx,single_game_col] + (df.at[idx,col] * (df.at[idx,'career_games_played'] - 1))) / df.at[idx,'career_games_played']
            elif col[0:7] == 'season_':
                df.at[idx,col] = (df.at[idx,single_game_col] + (df.at[idx,col] * (df.at[idx,'season_games_played'] - 1))) / df.at[idx,'season_games_played']
            else:
                raise Exception('Column not found, check what columns you\'re passing')
                
    return df
```

```python
updated = update_last_game(df)
```

```python
def get_upcoming_matches(matches, fixture, round_num):
    next_round = fixture[fixture['round']==str(round_num)]
    # Get list of home and away
    matches.sort_values(by=['season','round'],ascending=False,inplace=True)
    teams = list(next_round['hometeam'])
    teams = list(zip(teams,list(next_round['awayteam']))) # (home, away)
    return teams
```

```python
player_cols_to_agg = [
    'AFLfantasy',
    'Supercoach',
    'behinds',
    'bounces',
    'brownlow',
    'clangers',
    'clearances',
    'contested_marks',
    'contested_poss',
    'disposals',
    'frees_against',
    'frees_for',
    'goal_assists',
    'goals',
    'handballs',
    'hitouts',
    'inside50',
    'kicks',
    'marks',
    'marks_in_50',
    'one_percenters',
    'rebound50',
    'tackles',
    'tog',
    'uncontested_poss',
    'centre_clearances',
    'disposal_efficiency',
    'effective_disposals',
    'intercepts',
    'metres_gained',
    'stoppage_clearances',
    'score_involvements',
    'tackles_in_50',
    'turnovers',
    'matchid',
    'next_matchid',
    'team'
]
```

```python
from typing import List
def define_aggregate_columns(player_cols_to_agg: List) -> List:
    """
    Defines which columns we'd like to aggregate to the team level
    """
    player_cols_to_agg_temp = list(filter(
        lambda x: x not in ['matchid','next_matchid','team'],player_cols_to_agg))
    player_cols_to_agg.extend(
        ['career_' + col for col in player_cols_to_agg_temp])
    player_cols_to_agg.extend(
        ['season_' + col for col in player_cols_to_agg_temp])
    player_cols_to_agg.extend(['career_games_played','season_games_played'])
    return player_cols_to_agg
```

```python
player_cols = define_aggregate_columns(player_cols_to_agg)
```

```python
aggregate = players[player_cols].groupby(['next_matchid','team']).agg('mean')
```
